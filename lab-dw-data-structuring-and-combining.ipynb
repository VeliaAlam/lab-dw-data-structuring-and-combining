```python
import pandas as pd

url_1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url_2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url_3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

# DataFrame 1:
df_1 = pd.read_csv(url_1)

# DataFrame 2:
df_2 = pd.read_csv(url_2)

# DataFrame 3:
df_3 = pd.read_csv(url_3)
```

```python
display(df_1)
display(df_2)
display(df_3)
```

```python
print(df_1.columns)
print(df_2.columns)
print(df_3.columns)
```

```python
#Standardize column names
def standardize_columns(df):
    df = df.copy()
    df.columns = (
        df.columns.str.strip()                     
                  .str.lower()                   
                  .str.replace('state', 'st')     
                  .str.replace('gender', 'gender')  
    )
    return df


```

```python
df_1 = standardize_columns(df_1)
df_2 = standardize_columns(df_2)
df_3 = standardize_columns(df_3)
```

```python
common_cols = df_1.columns
df_2 = df_2[common_cols]
df_3 = df_3[common_cols]
```

```python
print(df_1.columns)
print(df_2.columns)
print(df_3.columns)
```

```python
combined_df = pd.concat([df_1, df_2, df_3], ignore_index=True)

display(combined_df)
```

```python
#Standardize the column names of the dataframe
combined_df.columns = (
    combined_df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace(r'\bst\b', 'state', regex=True)
    .str.replace('__', '_')
)
combined_df.head()
```

```python
# Clean 'gender' column
combined_df['gender'] = combined_df['gender'].str.strip().str.lower()
combined_df['gender'] = combined_df['gender'].replace({
    'f': 'F', 'female': 'F', 'femal': 'F',
    'm': 'M', 'male': 'M'
})

# Clean 'state' column
combined_df['state'] = combined_df['state'].replace({
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington'})

# Clean 'education' column
combined_df['education'] = combined_df['education'].replace({
    'bachelors': 'bachelor'
})

# Clean 'customer_lifetime_value' column (remove %)
combined_df['customer_lifetime_value'] = combined_df['customer_lifetime_value'].str.replace('%', '', regex=False)

# Clean 'vehicle_class' column
combined_df['vehicle_class'] = combined_df['vehicle_class'].replace({
    'Sports Car': 'luxury',
    'Luxury SUV': 'luxury',
    'Luxury Car': 'luxury'
})
```

```python
combined_df
```

```python
# Convert 'customer_lifetime_value' to numeric (ignore errors for NaN)

combined_df['customer_lifetime_value'] = pd.to_numeric(combined_df['customer_lifetime_value'], errors='coerce')

# Extract the middle value from 'number_of_open_complaints' and convert to numeric
def extract_middle(value):
    try:
        parts = str(value).split('/')
        if len(parts) == 3:
            return int(parts[1])
        else:
            return pd.NA
    except:
        return pd.NA

combined_df['number_of_open_complaints'] = combined_df['number_of_open_complaints'].apply(extract_middle)
combined_df['number_of_open_complaints'] = pd.to_numeric(combined_df['number_of_open_complaints'], errors='coerce')

print(combined_df['number_of_open_complaints'].unique())
print(combined_df['number_of_open_complaints'].dtype)
```

```python
combined_df.dtypes
```

```python
combined_df.isnull().sum()
```

```python
combined_df.dropna(thresh=2, inplace=True) #At least 2 no-nulls values 
combined_df
```

```python
print(combined_df.isnull().sum())
```

```python
# Numeric columns
numeric_columns = combined_df.select_dtypes(include='number').columns

# Categorical columns (usually strings)
categorical_columns = combined_df.select_dtypes(include='object').columns
```

```python
numeric_columns
```

```python
categorical_columns
```

```python
combined_df.describe()
```

```python
# Fill missing values in 'gender' with the most frequent value (mode)
combined_df['gender'].fillna(combined_df['gender'].mode()[0], inplace=True)

# Fill missing values in 'number_of_open_complaints' with 0
combined_df['number_of_open_complaints'].fillna(0, inplace=True)

# Drop the 'customer_lifetime_value' column due to high number of missing values
combined_df.drop(columns='customer_lifetime_value', inplace=True)
combined_df
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
#Create a pivot table to summarize the data
pivot_table = df.pivot_table(
    index='sales_channel',
    values='total_claim_amount',
    aggfunc='sum'
).round(2)
```

```python
pivot_table = pivot_table.sort_values(by='total_claim_amount', ascending=False)
```

```python
print(pivot_table)
pivot_table.to_csv('pivot_table.csv')
```

```python
"""
The Agent sales channel generated the highest total revenue, totaling approximately $1.81 million in claim amounts. It is followed by the Branch channel with around $1.3 million, and the Call Center with about $927,000. The Web channel generated the lowest total revenue at approximately $707,000.
This suggests that the Agent channel is the most effective at generating sales revenue. The Web channel, while important, currently contributes the least and may need further evaluation or improvement to increase its impact. The company might consider investing resources to support and optimize the higher-performing channels, especially the Agent channel.
"""
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
#Create a pivot table with number of open complaints, policy_type and month
pivot_complaints = df.pivot_table(
    index='month',
    columns='policy_type',
    values='number_of_open_complaints',
    aggfunc='sum',
    fill_value=0
)
print(pivot_complaints)
```

```python
#table pivot to large format
long_format = pivot_complaints.reset_index().melt(
    id_vars='month',
    var_name='policy_type',
    value_name='total_complaints'
)
print(long_format)
```
